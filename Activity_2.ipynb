```python
num = [10,20]
den = [1,15,50,0]
```

```python
import control as ctrl
import matplotlib.pyplot as plt

G = ctrl.tf(num, den)
```

```python
print (G)
```

```python
t, y = ctrl.step_response(G)
plt.title("Graph")
plt.plot(t, y)
plt.xlabel("Time (s)")
plt.ylabel("Output")
plt.grid()
plt.show()
```
