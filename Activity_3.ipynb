```python
import control as ctrl
import matplotlib.pyplot as plt

num = [5,5]
den = [1, 4, 3]
G = ctrl.TransferFunction(num, den)
print(G)
```

```python
poles = ctrl.poles(G)
print("Poles of the transfer function:", poles)
```

```python
zeros = ctrl.zeros(G)
print("Zeros of the transfer function:", zeros)
```

```python
ctrl.pzmap(G, grid=True)
plt.show()
```
