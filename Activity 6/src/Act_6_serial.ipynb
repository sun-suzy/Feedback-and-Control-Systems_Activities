```python
import serial
import time
```

```python
arduino = serial.Serial(port='COM13', baudrate=9600, 
timeout=1)
time.sleep(2) 
```

```python
while True:
    command = input("Type 1 to switch the LED ON, 0 to switch it OFF, or 'q' to exit:")

    if command == '1':
        arduino.write(b'1')
    elif command == '0':
        arduino.write(b'0')
    elif command ==  'q':
        break
    else:
        print("Invalid Command")

arduino.close()
```
