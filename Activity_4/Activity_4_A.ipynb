```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1] #numerator
den = [1, 2, 2] #denominator
G = ctrl.tf(num, den)
print(G)
```

```python
poles = ctrl.poles(G)
print("Poles:", poles)
```

```python
zero = ctrl.zeros(G)
print("Zero:", zero)
```

```python
import matplotlib.pyplot as plt

ctrl.pzmap(G, grid=True)
plt.show()
```

```python
t, y = ctrl.step_response(G)
plt.plot(t, y)
plt.title('Step Response of the System')
plt.xlabel('TIME (s)')
plt.ylabel('OUTPUT')
plt.grid()
plt.show()
```
