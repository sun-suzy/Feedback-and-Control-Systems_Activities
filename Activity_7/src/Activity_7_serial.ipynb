```python
import serial
import time

def start_motor_control():
    computer_port = 'COM13' 
    baud_rate = 9600
    
    try:
        arduino = serial.Serial(computer_port, baud_rate, timeout=1)
        time.sleep(2) # Wait for Arduino to reboot
        
        while True:
            input_value = input("\nDegree? (0-180) or 'quit': ").strip()
            if input_value.lower() == 'quit': break
            
            try:
                degree = int(input_value)
                if 0 <= degree <= 180:
                    arduino.write(f"{degree}\n".encode('utf-8'))
                    time.sleep(0.1)
                    if arduino.in_waiting > 0:
                        reply = arduino.readline().decode('utf-8').strip()
                        print(f"Update: {reply}")
                else:
                    print(f"Problem: {degree} is outside range.")
            except ValueError:
                print("Problem: Please enter a whole number.")
                
        arduino.close()
    except Exception as error:
        print(f"Serial Port Error: {error}")

if __name__ == "__main__":
    start_motor_control()
```
