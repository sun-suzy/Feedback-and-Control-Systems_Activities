```python
from langchain_openai import ChatOpenAI

# Connect to LM Studio
llm = ChatOpenAI(
    base_url="http://127.0.0.1:1234/v1", # Local address
    api_key="lm-studio",    # Any string works
    model="qwen/qwen3-4b",  # Model name
    temperature=0.7,        # Creativity level
    streaming=True          # Enable typing effect
)

print("AI connected!")
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

# 1. Setup prompt with a blank space for memory
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Use chat history to answer."),
    MessagesPlaceholder(variable_name="history"), # Memory
    ("user", "{input}")                           # User input
])

# 2. Link prompt and AI
chain = prompt | llm

# 3. Create memory storage
chat_memory = InMemoryChatMessageHistory()

# 4. Function to load memory
def get_chat_history(session_id: str):  
    return chat_memory

# 5. Combine AI, prompt, and memory
ai_with_memory = RunnableWithMessageHistory(
    chain,
    get_chat_history,
    input_messages_key="input",     # Link input
    history_messages_key="history", # Link memory
)

print("Memory ready!")
```

```python
print("Chat started! Type 'quit' to exit.\n" + "="*50)

while True:
    # Get user input (LOOK AT THE TOP OF YOUR VS CODE WINDOW!)
    my_question = input("Type your question here: ")
    
    # SAFETY CHECK: If you press enter without typing anything, ignore it!
    if not my_question.strip():
        continue
    
    # Check for exit
    if my_question.lower() in ['exit', 'quit']:
        print("Bye!")
        break
        
    # Print what you typed so it stays on the screen
    print(f"\nMe: {my_question}")
    print("AI: ", end="")
    
    # Send question to AI
    stream = ai_with_memory.stream(
        {"input": my_question},
        config={"configurable": {"session_id": "chat_1"}} # Chat ID
    )
    
    # Print response chunks
    for chunk in stream:
        if chunk.content:
            print(chunk.content, end="", flush=True)
            
    print("\n" + "-"*50)
```
